```python
import numpy as np
import pandas as pd
import ast
```

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
movies = movies.merge(credits, on = 'title')
```

```python
movies.head(2)
```

```python
movies = movies[['id', 'title','overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.head(2)
```

```python
movies.isnull().sum() #for finding out the null values in our data set
```

```python
movies.dropna(inplace = True)
```

```python
movies.duplicated().sum()
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
# movies.head(4)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter < 3:
            L.append(i['name'])
            counter +=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#movies['overview'][1]
```

```python
#movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.sample(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['genres'] = movies['genres'].apply(collapse)
movies['overview'] = movies['overview'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['id', 'title' , 'tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][1]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
# from nltk.stem.porter import PorterStemmer as ps

from nltk.stem import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

## To get autosuggested words just presss Tab key

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# vectors[0].dtype
vectors[0]
```

```python
cv.get_feature_names()

# In this we can see many names are repeated like activity and activities so we want them to be same so we will
# use a technique called steaming which will convert both this word into activity only
```

```python
new_df['tags'][0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommand(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])
    for i in movies_list[1:6]:
        print(new_df.iloc[i[0]].title)
```
